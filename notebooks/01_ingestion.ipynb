# 01 - Smart Meter Data Ingestion

## Overview
This notebook covers:
- Loading raw smart meter data from CSV
- Schema validation
- Initial data exploration
- Data quality assessment
- Summary statistics

## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
import logging
from datetime import datetime
import json

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

print("Imports successful")
```

## 2. Configure Paths

```python
# Setup project paths
PROJECT_ROOT = Path.cwd().parent if 'notebooks' in str(Path.cwd()) else Path.cwd()
DATA_RAW = PROJECT_ROOT / 'data' / 'raw'
DATA_PROCESSED = PROJECT_ROOT / 'data' / 'processed'

# Create directories
DATA_RAW.mkdir(parents=True, exist_ok=True)
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

print(f"Project Root: {PROJECT_ROOT}")
print(f"Raw Data Dir: {DATA_RAW}")
print(f"Processed Data Dir: {DATA_PROCESSED}")
```

## 3. Load Raw Data

```python
# Find and load the raw smart meter data
raw_file = DATA_RAW / 'raw_smart_meter.csv'

if not raw_file.exists():
    raise FileNotFoundError(f"Raw data file not found: {raw_file}")

print(f"Loading data from: {raw_file}")
df = pd.read_csv(raw_file)

print(f"\n[CHECK]“ Data loaded successfully!")
print(f"  Shape: {df.shape[0]:,} rows Ã— {df.shape[1]} columns")
print(f"  Memory: {df.memory_usage(deep=True).sum() / 1024 / 1024:.2f} MB")
```

## 4. Schema Validation

```python
# Display schema information
print("\n=== SCHEMA INFORMATION ===")
print(f"\nColumn Names: {list(df.columns)}")
print(f"\nData Types:\n{df.dtypes}")

# Schema validation
required_columns = ['Timestamp', 'Meter_ID', 'Zone', 'Active_Power_kW', 'Reactive_Power_kVAR']
missing_cols = [col for col in required_columns if col not in df.columns]

if missing_cols:
    print(f"\nâš  WARNING: Missing required columns: {missing_cols}")
else:
    print(f"\n[CHECK]“ All required columns present")
```

## 5. Initial Data Exploration

```python
# Display first records
print("\n=== FIRST 5 RECORDS ===")
print(df.head())

print("\n=== LAST 5 RECORDS ===")
print(df.tail())

print("\n=== BASIC INFO ===")
print(df.info())
```

## 6. Data Quality Assessment

```python
# Null values analysis
print("\n=== NULL VALUES ===")
null_counts = df.isnull().sum()
null_pct = (df.isnull().sum() / len(df) * 100).round(2)
null_df = pd.DataFrame({
    'Column': null_counts.index,
    'Null_Count': null_counts.values,
    'Null_Percentage': null_pct.values
}).sort_values('Null_Count', ascending=False)

print(null_df[null_df['Null_Count'] > 0])
if null_df['Null_Count'].sum() == 0:
    print("No null values found [CHECK]“")
```

```python
# Duplicate analysis
print("\n=== DUPLICATE RECORDS ===")
total_duplicates = df.duplicated().sum()
complete_duplicates = df.duplicated(subset=['Timestamp', 'Meter_ID']).sum()

print(f"Complete row duplicates: {total_duplicates}")
print(f"Meter-Timestamp duplicates: {complete_duplicates}")

if total_duplicates > 0:
    print(f"\nDuplicate percentage: {total_duplicates/len(df)*100:.2f}%")
else:
    print("[CHECK]“ No duplicates found")
```

## 7. Summary Statistics

```python
# Numerical columns statistics
print("\n=== NUMERICAL SUMMARY ===")
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(df[numeric_cols].describe().round(2))
```

```python
# Categorical columns summary
print("\n=== CATEGORICAL SUMMARY ===")
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

for col in categorical_cols:
    unique_count = df[col].nunique()
    print(f"\n{col}:")
    print(f"  Unique values: {unique_count}")
    if unique_count <= 10:
        print(f"  Value counts:\n{df[col].value_counts()}")
```

## 8. Time Series Analysis

```python
# Parse timestamp if needed
if 'Timestamp' in df.columns:
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
    
    print("\n=== TIME SERIES INFO ===")
    print(f"Date Range: {df['Timestamp'].min()} to {df['Timestamp'].max()}")
    print(f"Duration: {df['Timestamp'].max() - df['Timestamp'].min()}")
    print(f"Frequency: {df['Timestamp'].dtype}")
    print(f"\nRecords per day: {len(df) / ((df['Timestamp'].max() - df['Timestamp'].min()).days or 1):.0f}")
```

## 9. Meter and Zone Distribution

```python
# Meter distribution
if 'Meter_ID' in df.columns:
    print("\n=== METER DISTRIBUTION ===")
    print(f"Unique meters: {df['Meter_ID'].nunique()}")
    print(f"\nRecords per meter:")
    print(df['Meter_ID'].value_counts().head(10))

# Zone distribution
if 'Zone' in df.columns:
    print("\n=== ZONE DISTRIBUTION ===")
    print(f"Unique zones: {df['Zone'].nunique()}")
    print(f"\nZone breakdown:")
    print(df['Zone'].value_counts())
```

## 10. Generate Ingestion Report

```python
# Create ingestion report
ingestion_report = {
    'timestamp': datetime.now().isoformat(),
    'file_path': str(raw_file),
    'total_records': len(df),
    'total_columns': len(df.columns),
    'columns': list(df.columns),
    'data_types': df.dtypes.astype(str).to_dict(),
    'null_values': df.isnull().sum().to_dict(),
    'duplicate_records': int(df.duplicated().sum()),
    'memory_mb': float(df.memory_usage(deep=True).sum() / 1024 / 1024),
    'numeric_columns': numeric_cols,
    'categorical_columns': categorical_cols
}

print("\n=== INGESTION REPORT ===")
print(json.dumps(ingestion_report, indent=2, default=str))

# Save report
report_file = DATA_PROCESSED / 'ingestion_report.json'
with open(report_file, 'w') as f:
    json.dump(ingestion_report, f, indent=2, default=str)

print(f"\n[CHECK]“ Report saved to: {report_file}")
```
